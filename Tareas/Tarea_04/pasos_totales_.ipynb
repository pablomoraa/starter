```python
import pandas as pd
import matplotlib.pyplot as plt

workbook1 = "/content/drive/MyDrive/actividad_fisica_tabla.csv"

df = pd.read_csv(workbook1)
print(df.head())

valores = df[["Mes","SUM de pasos"]]
print(valores)

ax = valores.plot.bar(x="Mes", y = "SUM de pasos", rot=0)

plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = ("2019", "2020", "2021", "2022", "2023")
pasos = {"SUM pasos totales" : (183302, 80213, 58736, 195458, 283578)}

x = np.arange(len(years))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in pasos.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Suma de pasos')
ax.set_title('Cantidad de pasos por año')
ax.set_xticks(x, years)
ax.legend(loc='upper left', ncols=1)
ax.set_ylim(0, 350000)

plt.show()
```
